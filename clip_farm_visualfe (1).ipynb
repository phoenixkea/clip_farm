# measured the average brightness

measured the average brightness

```python
import cv2
import os
import numpy as np

video_files = [
    "#catsoftiktok.mp4",
    "500 pound jynxzi.mp4",
    "am i pretty? Danny Gonzalez Shorts.mp4",
    "cuando no encontras lugar.mp4",
    "Islam Makhachev VS Adin .mp4",
    "Product Link in Bio ( # 2300 ) ✅ 4-Layer Japan....mp4",
    "Sigma Kid Love #funny #sigma.mp4",
    "TASTING LITTLE MOONS.mp4",
    "When the toilet occupied! .mp4",
    "本以为胜券在握，没想到天都塌了#母女日常 #亲子日常#万万想不到.mp4"
]
brightness_results = []

for video_file in video_files:
    if not os.path.exists(video_file):
        print(f"Video file '{video_file}' not found.")
        continue

    cap = cv2.VideoCapture(video_file)

    frame_brightness = []
    frame_count = 0

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        # convert the frame to grayscale to measure brightness
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

        # compute average brightness of the frame
        avg_brightness = np.mean(gray)
        frame_brightness.append(avg_brightness)
        frame_count += 1

    cap.release()

    # compute the overall average brightness
    overall_avg_brightness = np.mean(frame_brightness)
    result = {
        "video_file": video_file,
        "total_frames": frame_count,
        "overall_avg_brightness": round(overall_avg_brightness, 2)
    }

    brightness_results.append(result)

# show results for all videos
for r in brightness_results:
    print(r)

print(f"Processed {len(brightness_results)} video(s)")
```

Why This Matters for the Clip Farm
Brightness can affect:

Emotional tone: Brighter videos often feel upbeat or clearer; dark videos may signal drama or mystery.

Attention span: High brightness and contrast can improve visual clarity and viewer retention.

Aesthetic trends: Viral clips often use high exposure or stylized lighting (especially on platforms like TikTok or Reels).

# motion intensity

For each frame, it compares it with the previous one.

It calculates how different the two frames are using absolute pixel difference.

The more different the frames are, the more motion is happening (like rapid camera movements, jump cuts, or lots of activity).

We average this value across the whole video for a single "motion intensity" score.

```python
video_files = [
    "#catsoftiktok.mp4",
    "500 pound jynxzi.mp4",
    "am i pretty? Danny Gonzalez Shorts.mp4",
    "cuando no encontras lugar.mp4",
    "Islam Makhachev VS Adin .mp4",
    "Product Link in Bio ( # 2300 ) ✅ 4-Layer Japan....mp4",
    "Sigma Kid Love #funny #sigma.mp4",
    "TASTING LITTLE MOONS.mp4",
    "When the toilet occupied! .mp4",
    "本以为胜券在握，没想到天都塌了#母女日常 #亲子日常#万万想不到.mp4"
]

motion_results = []

for video_file in video_files:
    if not os.path.exists(video_file):
        print(f"Video file '{video_file}' not found.")
        continue

    cap = cv2.VideoCapture(video_file)
    prev_frame = None
    motion_intensity = []

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

        if prev_frame is not None:
            # absolute difference between current and previous frame
            diff = cv2.absdiff(gray, prev_frame)
            mean_diff = np.mean(diff)  # intensity of motion between frames
            motion_intensity.append(mean_diff)

        prev_frame = gray

    cap.release()

    avg_motion = round(np.mean(motion_intensity), 2) if motion_intensity else 0

    result = {
        "video_file": video_file,
        "avg_motion_intensity": avg_motion
    }
    motion_results.append(result)

for r in motion_results:
    print(r)

print(f"Processed {len(motion_results)} video(s) for motion intensity.")
```

# video contrast

```python
video_files = [
    "#catsoftiktok.mp4",
    "500 pound jynxzi.mp4",
    "am i pretty? Danny Gonzalez Shorts.mp4",
    "cuando no encontras lugar.mp4",
    "Islam Makhachev VS Adin .mp4",
    "Product Link in Bio ( # 2300 ) ✅ 4-Layer Japan....mp4",
    "Sigma Kid Love #funny #sigma.mp4",
    "TASTING LITTLE MOONS.mp4",
    "When the toilet occupied! .mp4",
    "本以为胜券在握，没想到天都塌了#母女日常 #亲子日常#万万想不到.mp4"
]

contrast_results = []

for video_file in video_files:
    if not os.path.exists(video_file):
        print(f"Video file '{video_file}' not found.")
        continue

    cap = cv2.VideoCapture(video_file)
    frame_contrast = []
    frame_count = 0

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

        # calculate contrast for this frame (max - min)
        contrast = np.max(gray) - np.min(gray)
        frame_contrast.append(contrast)
        frame_count += 1

    cap.release()

    # average contrast across all frames
    avg_contrast = round(np.mean(frame_contrast), 2) if frame_contrast else 0

    result = {
        "video_file": video_file,
        "total_frames": frame_count,
        "avg_contrast": avg_contrast
    }

    contrast_results.append(result)

for r in contrast_results:
    print(r)

print(f"Processed {len(contrast_results)} video(s) for contrast.")
```

contrast=max pixel value−min pixel value

A reaction video with a bright background and strong shadows on the speaker’s face feels more intense.

A low-contrast video (e.g., soft lighting, fog, dull color) might feel dreamy or artsy — less viral, but aesthetic.

Many clips that go viral on TikTok use high contrast filters to make colors pop on phone screens.

✦ In Your Clip Farm Context:
If you're building a model to spot patterns in viral content, measuring average contrast can:

Help distinguish between calm vs intense visual tones.

Combine with motion intensity and brightness to better understand pacing and style.

Feed into a feature set for a model to learn visual aesthetics tied to virality.

```python
from fer import FER
import cv2
import os
```

```python
video_files = [
    "#catsoftiktok.mp4",
    "500 pound jynxzi.mp4",
    "am i pretty? Danny Gonzalez Shorts.mp4",
    "cuando no encontras lugar.mp4",
    "Islam Makhachev VS Adin .mp4",
    "Product Link in Bio ( # 2300 ) ✅ 4-Layer Japan....mp4",
    "Sigma Kid Love #funny #sigma.mp4",
    "TASTING LITTLE MOONS.mp4",
    "When the toilet occupied! .mp4",
    "本以为胜券在握，没想到天都塌了#母女日常 #亲子日常#万万想不到.mp4"
]

emotion_results = []

# initialize fer detector
emotion_detector = FER()

for video_file in video_files:
    if not os.path.exists(video_file):
        print(f"Video file '{video_file}' not found.")
        continue

    cap = cv2.VideoCapture(video_file)
    emotion_log = []
    frame_count = 0

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        result = emotion_detector.detect_emotions(frame)
        if result:
            # result[0] refers to first face detected
            emotion_log.append(result[0]['emotions'])

        frame_count += 1

    cap.release()

    # aggregate emotions across frames
    if emotion_log:
        # average each emotion score across all frames
        emotion_sums = {k: 0 for k in emotion_log[0]}
        for e in emotion_log:
            for k in e:
                emotion_sums[k] += e[k]

        num_frames = len(emotion_log)
        avg_emotions = {k: round(v / num_frames, 2) for k, v in emotion_sums.items()}
    else:
        avg_emotions = {}

    result = {
        "video_file": video_file,
        "total_frames_with_faces": len(emotion_log),
        "avg_emotions": avg_emotions
    }
    emotion_results.append(result)

# output results
for r in emotion_results:
    print(r)

print(f"Processed {len(emotion_results)} video(s) for emotional expression.")
```

# detecting cut scens, messures avg shot duration

```python
import cv2
import os
import numpy as np
```

```python
video_files = [
    "#catsoftiktok.mp4",
    "500 pound jynxzi.mp4",
    "am i pretty? Danny Gonzalez Shorts.mp4",
    "cuando no encontras lugar.mp4",
    "Islam Makhachev VS Adin .mp4",
    "Product Link in Bio ( # 2300 ) ✅ 4-Layer Japan....mp4",
    "Sigma Kid Love #funny #sigma.mp4",
    "TASTING LITTLE MOONS.mp4",
    "When the toilet occupied! .mp4",
    "本以为胜券在握，没想到天都塌了#母女日常 #亲子日常#万万想不到.mp4"
]

shot_length_results = []

for video_file in video_files:
    if not os.path.exists(video_file):
        print(f"Video file '{video_file}' not found.")
        continue

    cap = cv2.VideoCapture(video_file)
    prev_frame = None
    threshold = 30  # sensitivity for detecting scene cuts
    last_cut_frame = 0
    frame_count = 0
    shot_lengths = []

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        if prev_frame is not None:
            diff = cv2.absdiff(gray, prev_frame)
            mean_diff = np.mean(diff)
            if mean_diff > threshold:
                shot_length = frame_count - last_cut_frame
                shot_lengths.append(shot_length)
                last_cut_frame = frame_count

        prev_frame = gray
        frame_count += 1

    cap.release()

    if shot_lengths:
        avg_shot = round(np.mean(shot_lengths), 2)
        shot_length_results.append({
            "video_file": video_file,
            "avg_shot_length_frames": avg_shot,
            "total_shots": len(shot_lengths)
        })
    else:
        shot_length_results.append({
            "video_file": video_file,
            "avg_shot_length_frames": 0,
            "total_shots": 0
        })

for r in shot_length_results:
    print(r)
```

# avg saturation + domanite colour

```python
import cv2
import numpy as np
import os
from sklearn.cluster import KMeans

video_files = [
    "#catsoftiktok.mp4",
    "500 pound jynxzi.mp4",
    "am i pretty? Danny Gonzalez Shorts.mp4",
    "cuando no encontras lugar.mp4",
    "Islam Makhachev VS Adin .mp4",
    "Product Link in Bio ( # 2300 ) ✅ 4-Layer Japan....mp4",
    "Sigma Kid Love #funny #sigma.mp4",
    "TASTING LITTLE MOONS.mp4",
    "When the toilet occupied! .mp4",
    "本以为胜券在握，没想到天都塌了#母女日常 #亲子日常#万万想不到.mp4"
]

color_results = []
saturation_data = []
color_data = []

for video_file in video_files:
    if not os.path.exists(video_file):
        print(f"Video file '{video_file}' not found.")
        continue

    cap = cv2.VideoCapture(video_file)
    saturation_values = []
    dominant_colors = []

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        #cConvert to HSV to get saturation
        hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
        saturation = np.mean(hsv[:, :, 1])  # Extract S channel
        saturation_values.append(saturation)

        # resize frame to reduce processing time for KMeans
        small_frame = cv2.resize(frame, (64, 64))
        pixels = small_frame.reshape(-1, 3)

        # use KMeans to get dominant colors
        kmeans = KMeans(n_clusters=3, n_init=10)
        kmeans.fit(pixels)
        dominant_rgb = kmeans.cluster_centers_.mean(axis=0)
        dominant_colors.append(dominant_rgb)

    cap.release()

    avg_saturation = round(np.mean(saturation_values), 2)
    avg_dominant_color = tuple(map(int, np.mean(dominant_colors, axis=0)))

    saturation_data.append({
        "video_file": video_file,
        "avg_saturation": avg_saturation
    })
    color_data.append({
        "video_file": video_file,
        "avg_dominant_color_rgb": avg_dominant_color # storing as tuple/list of integers
    })

    r, g, b = avg_dominant_color
color_data.append({
    "video_file": video_file,
    "dominant_color_r": r,
    "dominant_color_g": g,
    "dominant_color_b": b
})


for r in color_results:
    print(r)
```

# audio

Use moviepy to extract the audio from each video.

Use librosa to calculate how “loud” the video is (volume analysis).


High loud_ratio → More action, music, hype energy (good for virality)

Low loud_ratio → Calm, slow, quiet — possibly serious or emotional tone

```python
from moviepy.editor import VideoFileClip
import librosa
import numpy as np
import os
```

```python


video_files = [
    "#catsoftiktok.mp4",
    "500 pound jynxzi.mp4",
    "am i pretty? Danny Gonzalez Shorts.mp4",
    "cuando no encontras lugar.mp4",
    "Islam Makhachev VS Adin .mp4",
    "Product Link in Bio ( # 2300 ) ✅ 4-Layer Japan....mp4",
    "Sigma Kid Love #funny #sigma.mp4",
    "TASTING LITTLE MOONS.mp4",
    "When the toilet occupied! .mp4",
    "本以为胜券在握，没想到天都塌了#母女日常 #亲子日常#万万想不到.mp4"
]

audio_results = []

for video_file in video_files:
    if not os.path.exists(video_file):
        print(f"Video file '{video_file}' not found.")
        continue

    print(f"Processing audio from: {video_file}")

    # Extract and save audio
    try:
        clip = VideoFileClip(video_file)
        audio_path = "temp_audio.wav"
        clip.audio.write_audiofile(audio_path, verbose=False, logger=None)

        # Load audio using librosa
        y, sr = librosa.load(audio_path)

        # Calculate short-term energy (volume)
        frame_length = 2048
        hop_length = 512
        energy = np.array([
            sum(abs(y[i:i+frame_length]**2))
            for i in range(0, len(y), hop_length)
        ])

        # Determine loud sections
        volume_threshold = np.percentile(energy, 75)
        loud_frames = energy > volume_threshold
        loud_ratio = round(np.sum(loud_frames) / len(energy), 2)

        audio_results.append({
            "video_file": video_file,
            "loud_ratio": loud_ratio
        })

        os.remove(audio_path)  # Clean up temp file

    except Exception as e:
        print(f"Error processing {video_file}: {e}")

# Output results
for r in audio_results:
      print(r)
```
